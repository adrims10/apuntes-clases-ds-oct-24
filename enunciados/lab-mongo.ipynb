### 3. Todas las empresas fundadas entre 2000 y 2005, incluyendo ambos años. Recupera solo los campos `name` y `founded_year`.


```python
cond_precio = {"founded_year": {"$gte": 2000}}
cond_camas = {"founded_year": {'$lte':2005}}

# realizamos la query
find_and = pd.DataFrame(coleccion.find({"$and": [cond_precio, cond_camas]},{'name':1, '_id':0,'founded_year':1}))
find_and.head(2)
```
